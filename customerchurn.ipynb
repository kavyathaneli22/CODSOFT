```python
#importing the important libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
#loading the dataset
churn_detection = pd.read_csv('Churn_Modelling.csv')
```

```python
#Labelling the categorical data
churn_detection['Geography'].value_counts()
churn_detection.loc[churn_detection['Geography'] == 'France' ,'Geography'] = 0
churn_detection.loc[churn_detection['Geography'] == 'Germany' ,'Geography'] = 1
churn_detection.loc[churn_detection['Geography'] == 'Spain' ,'Geography'] = 2
```

```python
#Gender
churn_detection.loc[churn_detection['Gender'] == 'Female' ,'Gender'] = 0
churn_detection.loc[churn_detection['Gender'] == 'Male' ,'Gender'] = 1
```

```python
churn_detection
```

```python
#seperating the target variable
X = churn_detection.drop(['Exited'],axis = 1)
Y = churn_detection['Exited']
```

```python
#printing x
print(X)
```

```python
#printing Y
print(Y)
```

```python
#Splitting the data into training data and testing data
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=40)
print(X.shape,X_train.shape,X_test.shape)
```

```python
#creating the model
model = RandomForestClassifier(n_estimators=100, random_state=40)
```

```python
#Training the model
model.fit(X_train, y_train)
```

```python
#prediction and Model Evaluation
predict = model.predict(X_test)
print("Accuracy Score:", accuracy_score(y_test,predict))
```
